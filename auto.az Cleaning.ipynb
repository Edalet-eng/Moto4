```python
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
import pandas as pd
import numpy as np 
import datetime
import warnings
```

```python
pd.set_option('display.max_columns', None)
warnings.filterwarnings(action = 'ignore')
```

```python
filepath = 'auto2.csv'
df = pd.read_csv(filepath)
df.head()
```

```python
df = df.drop_duplicates()
```

```python
df.head()
```

```python
df.columns = df.columns.str.replace(pat = ' ', repl = '_')
df.columns = df.columns.str.lower()
df.columns = df.columns.str.strip()
df.head()
```

```python
def equate(df = None):
    df.sürətlər_qutusu = df.sürətlər_qutusu.str.replace(pat = 'avtomatik', repl = 'avtomat')
    df.ban_növü = df.ban_növü.str.replace(pat = 'ofrouder' , repl = 'offroader / suv')
    df.ban_növü = df.ban_növü.str.replace(pat = 'xetçbek' , repl = 'hetçbek')
    df.ban_növü = df.ban_növü.str.replace(pat = 'miniven/mikroavtobus' , repl = 'minivan')
    df.ban_növü = df.ban_növü.str.replace(pat = 'mototexnika' , repl = 'motosiklet')
    df.ban_növü = df.ban_növü.str.replace(pat = 'xüsusi texnika' , repl = 'yük maşını')
    df.rəng = df.rəng.apply(lambda x: x.replace(' (metallik)',''))
    return df
```

```python
def convert_currency(df = None):
    df['qiymet'] = df['qiymet'].apply(lambda x: int(x.replace('$',''))*1.7 if '$' in x else (int(x.replace('€',''))*1.81 if '€' in x else int(x.replace('AZN',''))))
    return df
```

```python
def convert_int(df = None):
    df.mühərrik_gucu = df.mühərrik_gucu.str.replace(pat = 'a.g.', repl = '')
    df.yürüş = df.yürüş.str.replace(pat = 'km', repl = '')
    df.yürüş = df.yürüş.str.replace(pat = ' ', repl = '')
    df.mühərrik_hecmi = pd.to_numeric(arg = df.mühərrik_hecmi, downcast = 'float')
    df[['mühərrik_gucu', 'yürüş']] = df[['mühərrik_gucu', 'yürüş']].applymap(func = lambda x: int(x))
    return df
```

```python
def convert_lower_case(df = None):
    df_obj = df.select_dtypes(include = 'object') 
    df[df_obj.columns] = df_obj.applymap(lambda x: np.nan if pd.isna(x) else x.lower())
    return df
```

```python
def create_new_columns(df = None):
    df['vuruğu_var'] = df.veziyyeti.apply(lambda x: 'vuruğu var' if ' vurulmuş ' in x else 'vuruğu yoxdur')
    df['rənglənib'] = df.veziyyeti.apply(lambda x: 'rənglənib' if ' orta ' in x else 'rənglənməyib')
    df['lehimli_disk'] = df.elave_techizat.apply(lambda x: np.nan if pd.isna(x) else 'var' if 'yüngül lehimli disklər' in x else 'yox')
    df['abs'] = df.elave_techizat.apply(lambda x: np.nan if pd.isna(x) else 'var' if 'abs' in x else 'yox')
    df['lyuk'] = df.elave_techizat.apply(lambda x: np.nan if pd.isna(x) else 'var' if 'lyuk' in x else 'yox')
    df['yağış_sensoru'] = df.elave_techizat.apply(lambda x: np.nan if pd.isna(x) else 'var' if 'yağış sensoru' in x else 'yox')
    df['mərkəzi_qapanma'] = df.elave_techizat.apply(lambda x: np.nan if pd.isna(x) else 'var' if 'mərkəzi qapanma' in x else 'yox')
    df['park_radarı'] = df.elave_techizat.apply(lambda x: np.nan if pd.isna(x) else 'var' if 'park radarı' in x else 'yox')
    df['kondisioner'] = df.elave_techizat.apply(lambda x: np.nan if pd.isna(x) else 'var' if 'kondisioner' in x else 'yox')
    df['oturacaqların_isidilməsi'] = df.elave_techizat.apply(lambda x: np.nan if pd.isna(x) else 'var' if 'qızdırılan oturacaqlar' in x else 'yox')
    df['dəri_salon'] = df.elave_techizat.apply(lambda x: np.nan if pd.isna(x) else 'var' if 'dəri salon' in x else 'yox')
    df['ksenon_lampalar'] = df.elave_techizat.apply(lambda x: np.nan if pd.isna(x) else 'var' if 'ksenon lampalar' in x else 'yox')
    df['arxa_görüntü_kamerası'] = df.elave_techizat.apply(lambda x: np.nan )
    df['yan_pərdələr'] = df.elave_techizat.apply(lambda x: np.nan )
    df['oturacaqların_ventilyasiyası'] = df.elave_techizat.apply(lambda x: np.nan)
    return df
```

```python
df = df.pipe(func = convert_currency).pipe(func = convert_int).pipe(func = convert_lower_case).pipe(func = equate).pipe(func = create_new_columns)
df.head()
```

```python
df = df.drop(columns = ['veziyyeti','tesvir','elave_techizat'])
```
